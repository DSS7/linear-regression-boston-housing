```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# Organize the dataset (without the target column)
df = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
```

```python
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]  # Extract the target (housing prices)
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
columns = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]
df = pd.DataFrame(data, columns=columns)
```

```python
df
```

```python
df.head()
```

```python
X = df  
y = target 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
X_train_scaled = scalar.fit_transform(X_train)
```

```python
X_test_scaled = scalar.transform(X_test)
```

```python
scalar.inverse_transform(X_train) #If we want to undo it we can use this, but here it isn't necessary.
```

## Here we have multiple input features. 
## So it becomes, MULTI LINEAR REGRESSION PROBLEM.

#### You use fit_transform on the training data to make the data consistent. Here, the computer learns how to adjust the data (e.g., finding the average or the highest/lowest values) and then changes the data based on what it learned.
#### You only use transform on the test data because you don’t want the computer to learn again—just apply what it learned from the training data.

```python
from sklearn.model_selection import cross_val_score
```

```python
model = LinearRegression()
model.fit(X_train_scaled,y_train)
```

```python
neg_mse = cross_val_score(model, X_train_scaled, y_train, cv=10, scoring='neg_mean_squared_error') 
```

```python
np.mean(neg_mse)
```

```python
print(f"Min target value: {y_train.min()}")
print(f"Max target value: {y_train.max()}")
print(f"Mean target value: {y_train.mean()}")
```

```python
model_predict = model.predict(X_test_scaled)
```

```python
model_predict
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, model_predict)
print(f"MSE: {mse}")
```

```python
rmse = np.sqrt(mse)
print(f"RMSE: {rmse}")
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, model_predict)
print(f"R² Score: {r2}") # Coefficient of Determination
```

```python
import seaborn as sns
```

```python
# Assuming model_predict and y_test are defined and contain your model's predictions and actual values
residuals = model_predict - y_test

# Create a distribution plot of the residuals
sns.displot(residuals, bins=30, kde=True)  # kde=True adds a kernel density estimate line

# Add labels and title
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.axvline(0, color='red', linestyle='--')  # Optional: Add a line at zero for reference
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regressor = Ridge()
```

```python
parameters = {"alpha":[0,1,2,5,10,11,15,16,17,18,19,20,30,40,50]}
ridge_cv = GridSearchCV(ridge_regressor,parameters,scoring="neg_mean_squared_error",cv=5)
ridge_cv.fit(X_train_scaled,y_train)
```

```python
print(ridge_cv.best_params_)
```

```python
print(ridge_cv.best_score_)
```

```python
ridge_pred = ridge_cv.predict(X_test_scaled)
```

```python
residuals = ridge_pred - y_test

# Create a distribution plot of the residuals
sns.displot(residuals, bins=30, kde=True)  # kde=True adds a kernel density estimate line

# Add labels and title
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Residuals')
plt.axvline(0, color='red', linestyle='--')  # Optional: Add a line at zero for reference
plt.show()
```

```python
r2 = r2_score(y_test, ridge_pred)
print(f"R² Score: {r2}") # Coefficient of Determination
```

```python
from sklearn.linear_model import Lasso
```

```python
print(lasso_cv.best_score_)
```
